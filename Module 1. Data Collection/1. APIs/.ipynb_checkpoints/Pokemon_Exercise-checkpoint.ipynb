```python
import requests
psy = requests.get('https://pokeapi.co/api/v2/type/psychic')
```

```python
psy.json()['pokemon']
```

```python
psy.json()['pokemon'][0]['pokemon']
```

```python
psy.json().keys()
```

```python
# Create a dictionary with empty lists that we will populate
psychic_list={'Number':[], 'Name':[], 'Link':[]}
for pokemon in psy.json()['pokemon']:
    pok_dict = pokemon['pokemon']
    link = pok_dict['url']
    name = pok_dict['name']
    number = int(link.split('/')[-2])
    #print(number, link)
    
    if number < 152:
        psychic_list['Number'].append(number)
        psychic_list['Name'].append(name)
        psychic_list['Link'].append(link)
    else:
        break
```
